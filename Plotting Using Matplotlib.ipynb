# Line Plot
.Bivariate a simple function 

.categorical -> numerical and numerical -> numerical

.Use case -Time Series data

```python
price=[48000,54000,57000,49000,47000,45000,45000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
```

```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman.head()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#label title
#colors (hex)
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#linewidth and style
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2)
plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#marker (size)
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#legend(loc)
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
```

```python
#limiting axes
price=[48000,54000,57000,49000,47000,45000,450000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
#grid
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
```

```python
#show
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
#plt.scatter on pandas data
df=pd.read_csv('batter.csv')
df=df.head(50)
df
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='+')
plt.title(' Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

# Bar chart
Bivariante Analysis

Numerical Vs Categorical

Use case-> Aggregate analysis of groups

```python
#Simple Bar chart
children=[10,20,40,10,30]
colors=['red','blue','green','pink','yellow']
plt.bar(colors,children,color='orange')
```

```python
#horizontal bar chart
plt.barh(colors,children,color='orange')
```

```python
#color and labels
df=pd.read_csv('batsman_season_record.csv')
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
#multiple bar chart
#xticks
df=pd.read_csv('batsman_season_record.csv')
plt.bar(np.arange(df.shape[0])-0.2 , df['2015'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]) , df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0]) +0.2, df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
#a problem
children=[10,20,40,10,30]
colors=['red red red red','blue blue blue','green green','pink pink','yellow yellow v yellow']
plt.bar(colors,children)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
#Stacked Bar Chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),label='2015')
plt.legend()
plt.show()
```

# Pie Chart

```python
#Univariate/Bivariate Analysis
#Categorical vs numerical
#Use Case-> To find contribution on a standard scale
```

```python
data=[20,34,49,80]
subjects=['eng','hindi','maths','science']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset
df=pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#percentage and color
#explode and shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0.8,0,0.1],shadow=True)
plt.show()
```
